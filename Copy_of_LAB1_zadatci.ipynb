### Zadatak 1. Cirkularna dvostruko vezana lista

U sklopu predavanja i vježbi implementirali smo jednostruko vezanu listu (engl. *singly-linked list*; SLL). Temeljem SLL implementacije (klase `Node` i `SLL_class`) te gradiva obrađenog u sklopu predavanja, vaš je zadatak implementirati i testirati **cirkularnu dvostruko vezanu listu** (engl. *circular doubly-linked list*). Cirkularna dvostruko vezana lista je struktura podataka u kojoj svaki čvor ima referencu na prethodni i sljedeći čvor u listi, pri čemu posljednji čvor pokazuje na prvi čvor (i obratno), što omogućuje efikasan pristup i manipulaciju podacima u bilo kojem smjeru unutar liste.

U sljedećem odjeljku koda nadopunite klase `Node` i `CDLL` tako da se ispunjava funkcionalnost cirkularne dvostruko vezane liste. Atribut `tail` nije potreban, već je dovoljno da lista ima samo *glavu*, odnosno atribut `head`.

````python
class Node:
    def __init__(self, data):
      self.value = data                   #u cirkularnoj listi se podaci referenciraju na iduci i prethodni zato next i prev
      self.next = None
      self.prev = None
                                          #return je placeholder dok se ne napise neki kod, u ovoj funkciji nije potrebna
class CDLL:
    def __init__(self):
        self.head = None
        self.size = 0

    def add_first(self, new_data):
        new_node = Node(new_data)
        if self.head is None:
          self.head = new_node
          new_node.next = new_node        #može i new.node.next = self.head, self.head i new_node pokazuju na istu stvar
          new_node.prev = new_node
        else:
          last = self.head.prev           #spremamo referencu na zadnjeg
          new_node.next = self.head       #novi čvor pokazuje na tadašnju glavu
          new_node.prev = last            #novi čvor pokazuje na onog tko je do sada bio zadnji
          last.next = new_node            #zadnji pokazuje na novi Node
          self.head.prev = new_node       #prošli od starog heada neće biti više zadnji nego će biti novi Node
          self.head = new_node            #novi postaje glava
        self.size += 1

    def add_last(self, new_data):
        new_node = Node(new_data)
        if self.head is None:
          self.head = new_node
          new_node.next = new_node
          new_node.prev = new_node
        else:
          last = self.head.prev
          new_node.prev = last
          new_node.next = self.head
          last.next = new_node
          self.head.prev = new_node       #node postaje prijašnji od tadašnjeg heada
        self.size += 1


    def delete_first(self):
        if self.head is None:
          print("Ova lista je prazna.")
        elif self.size == 1:
          self.head = None
          self.size -= 1
        else:
          last = self.head.prev
          next_node = self.head.next
          next_node.prev = last
          last.next = self.head.next
          self.head = next_node
          self.size -= 1

    def delete_last(self):
        if self.head is None:
          print("OVa lista je prazna.")

        elif self.size == 1:
          self.head = None
          self.size -= 1

        else:                             #da bi izbacili zadnji, od heada ćemo izbaciti prošli od prijašnjeg heada i smjernice oko njega prebrikati
          last = self.head.prev
          next_od_deletea = self.head
          next_od_deletea.prev = last.prev
          last.prev.next = self.head
          self.size -= 1

    def update_at_position(self, value, position):
        if position < 0 or position >= self.size:
          raise IndexError("Index out of range!")
        else:
          temp = self.head                #trenutni položaj je HEAD
          for i in range(position):       #za >position< puta će se head pomaknuti za idući
            temp = temp.next              #pomicanje heada na idući
          temp.value = value              #nakon što dođe to te pozicije tada njegova vrijednost postaje zadana vrijednost

    def delete_at_position(self, position):
        if position < 0 or position >= self.size:
          raise IndexError("Index out of range!")
        elif position == 0:               #prijedlog da ako se prvi obriše da head stavimo na idućeg, već imamo funkciju za to pa ću je pozvati
          self.delete_first()
        elif position == self.size - 1:
          self.delete_last()
        else:
          temp = self.head
          for i in range(position):
            temp = temp.next
    #-------------------------------------isti princip kod brisanja samo sada temp glumi HEAD i head ne mijenjamo
          temp.prev.next = temp.next
          temp.next.prazna = temp.prev
          self.size -= 1


    def insert_at_position(self, value, position):
        if position < 0 or position >= self.size:
          raise IndexError("Index out of range!")
        if position == 0:                                                       #gledamo ako je prva pozicija i iskoristimo funkciju
          self.add_first(value)
        elif position == self.size:                                             #gledamo ako je zadnja pozicija i koristimo već definiranu funkciju
          self.add_last(value)
        else:
          new_node = Node(value)
          temp = self.head
          for i in range(position-1):
            temp = temp.next
          new_node.next = temp.next                                             #prijašnji od novog nodea će biti isti od prošlog
          new_node.prev = temp                                                  #idući od novog nodea će biti taj nađeni
          temp.next.prev = new_node                                             #idući od sačuvanog će pokazivati na novi node
          temp.next = new_node                                                  #prijašnji od sačuvanog će pokazivati na novi node
          self.size += 1


    def rotate(self, steps=1):
        if self.head is None or self.size < 2 or steps == 0:
          return
        if steps > 0:
          for i in range(steps):
            self.head = self.head.next
        elif steps < 0:
          for i in range(abs(steps)):
            self.head = self.head.prev

    def __str__(self):
        temp = self.head
        out = ''
        while temp:
            out += str(temp.value) + ' '
            temp = temp.next
            if temp == self.head:
                break
        return out
````

### Zadatak 2. Simulacija *round-robin* raspoređivanja

*Round-robin* raspoređivanje (engl. *round-robin scheduling*) je algoritam raspoređivanja koji se koristi u operacijskim sustavima za upravljanje resursima procesora (CPU-a), posebno u višekorisničkim i višeprogramskim okruženjima. Ovaj jednostavan algoritam raspoređivanja dodjeljuje procesorsko vrijeme svakom procesu redom, bez obzira na prioritete ili dužinu izvršavanja procesa. Ključna komponenta *round-robin* raspoređivanja je cirkularni (rotirajući) raspored između procesa. Procesi se izvršavaju jedan po jedan, a kada jedan proces potroši svoj kvant vremena (engl. *time slice*) ili završi izvršavanje, prelazi se na sljedeći proces.

Prednosti round-robin raspoređivanja uključuju jednostavnost implementacije, fer raspoređivanje resursa, i prilagodljivost veličine kvanta vremena. Međutim, može imati i nedostatke u performansama, posebno u situacijama gdje postoje procesi s različitim zahtjevima za procesorskim vremenom ili kada se procesi često prekidaju i ponovno pokreću, što može dovesti do trošenja procesorskog vremena na prebacivanje između procesa.

U sljedećemo djeljku koda, koristeći jedan ili više objekata iz klase cirkularne liste `CDLL`, koju ste implementirali u prvom zadatku, ostvarite simulaciju scenarija koji uključuje:
* 5 procesa, čiji su nazivi spremljeni u Python listu `process_names`, a ukupna vremena izvođenja svakog procesa u listu `process_times` (u $\mu$s)
* Kvant vremena koji se dodjeljuje svakom procesu u trajanju od 100$\mu$s.
* Ispisivanje preostalog vremena do kraja izvođenja pojedinog procesa, za sve korake simulacije (ispisuje se početno vrijeme, prije dodjeljivanja CPU-a za izvođenje procesa).
* Ispisivanje odgovarajuće poruke u trenutku kada je proces gotov i kada ga treba izbaciti iz simulacije (odnosno izbaciti iz cirkularne liste).

Napomena: koristite metodu `rotate()`, tako da uvijek "glava" cirkularne liste predstavlja proces koji se upravo izvodi. Možete koristiti jednu listu u koju ćete spremiti imena procesa, a drugu listu u koju ćete spremiti preostala vremena.

Ispravan ispis bi trebao izgledati ovako:
```
Process A: 363
Process B: 120
Process C: 200
Process D: 232
Process E: 75
Process E finished!
Process A: 263
Process B: 20
Process B finished!
Process C: 100
Process C finished!
Process D: 132
Process A: 163
Process D: 32
Process D finished!
Process A: 63
Process A finished!

````python
process_names = ['A', 'B', 'C', 'D', 'E']
process_times = [363, 120, 200, 232, 75]
time_slice = 100

def round_robin(names, times, time_slicing):
  namesTimes = CDLL()
  for i in range(len(names)):
    namesTimes.add_last([names[i], times[i]])

  while namesTimes.head is not None:
    print(f"Proces {namesTimes.head.value[0]}: {namesTimes.head.value[1]}")

    namesTimes.head.value[1] -= time_slicing

    if namesTimes.head.value[1] <= 0:
      print(f"Proces {namesTimes.head.value[0]} finished!")
      namesTimes.delete_first()
    else:
      namesTimes.rotate()

round_robin(process_names, process_times, time_slice)
````

### Zadatak 3. Sklonište za životinje
Sklonište za životinje (engl. *animal shelter*) je standardni problem iz računarstva koji se često može naći u intervjuima za posao, a ispituje poznavanje redova kao apstraktne strukture podataka. Sklonište za životinje funkcionira na striktnom *first in, first out* principu, što znači da je ona životinja koja najdulje boravi u skloništu prva na redu za udomljavanje. Ako žele, udomitelji imaju pravo birati vrstu životinje (pas ili mačka). Implementirajte strukturu podataka (klasu) `AnimalShelter` koja omogućuje upravljanje redom životinja koje čekaju na udomljavanje. Struktura podataka treba podržavati sljedeće operacije:
* `enqueue`: Dodaje životinju (bilo psa ili mačku) u sklonište, označavajući ime životinje i vrstu životinje.
* `dequeue_any`: Uklanja najstariju životinju (bilo psa ili mačku) iz skloništa.
* `dequeue_dog`: Uklanja najstarijeg psa iz skloništa.
* `dequeue_cat`: Uklanja najstariju mačku iz skloništa.
Ključna ideja za rješavanje ovog problema efikasno je održavanje dva odvojena reda: jednog za pse i jednog za mačke. Potrebno je pratiti vremena dolaska svake pojedine životinje kako bi se uvijek mogla udomiti ona koja je prva došla, neovisno o tome radi li se o mački ili psu.

Vaš zadatak je implementirati ovu strukturu podataka u Python-u, uz napomenu da implementirate posebnu klasu `Animal` koja će sadržavati sve potrebne atribute (ime, vrsta, vrijeme dolaska). Objekte iz te klase možete spremati u Python strukturu `list` koju trebate koristiti u svrhu reda (*queue*-a).

````python
class Queue():
  def __init__(self):
    self.cdll = CDLL()

  def enqueue(self, value):
    self.cdll.add_last(value)

  def dequeue(self):
    if self.isEmpty():
      return None
    animal_to_return = self.cdll.head.value
    self.cdll.delete_first()
    return animal_to_return

  def peek(self):
    if not self.isEmpty():
      return self.cdll.head.value
    else:
      return None

  def isEmpty(self):
    return self.cdll.head is None

class Animal():
  position = 0
  def __init__(self, name, typeOfAnimal):
    self.name = name
    self.typeOfAnimal = typeOfAnimal
    self.position = Animal.position
    Animal.position += 1

class AnimalShelter():
  def __init__(self):
    self.dogs = Queue()
    self.cats = Queue()

  def enqueue(self, name, typeOfAnimal):
    animal = Animal(name, typeOfAnimal)
    if animal.typeOfAnimal == "Dog":
      self.dogs.enqueue(animal)
    elif animal.typeOfAnimal == "Cat":
      self.cats.enqueue(animal)
    else:
      print("Animal not supported.")

  def dequeue_any(self):
    if self.dogs.isEmpty() and self.cats.isEmpty():
      print("Nemamo životinja na stanju.")
      return
    elif self.dogs.isEmpty():
      return self.cats.dequeue()
    elif self.cats.isEmpty():
      return self.dogs.dequeue()
    else:
      dog = self.dogs.peek()
      cat = self.cats.peek()
      if dog is None:
        return self.cats.dequeue()
      if cat is None:
        return self.dogs.dequeue()
      if dog.position < cat.position:
        return self.dogs.dequeue()
      else:
        return self.dogs.dequeue()

  def dequeue_dog(self):
      if self.dogs.isEmpty():
          print("Nema pasa na stanju")
          return None
      return self.dogs.dequeue()

  def dequeue_cat(self):
      if self.cats.isEmpty():
          print("Nema mačaka na stanju..")
          return None
      return self.cats.dequeue()



````

### Zadatak 4. (Bonus) Josephusov problem

Flavije Josip bio je židovski vojskovođa u ratu protiv Rimljana.
Prema legendi, on i skupina od **n vojnika** bili su opkoljeni u pećini. Kako se nisu htjeli predati, odlučili su se ubiti međusobno po sljedećem pravilu:
Svi vojnici stanu u **krug** i počnu brojati. Svaki **k-ti vojnik** biva eliminiran, nakon čega se brojanje nastavlja od sljedećeg preživjelog. Postupak se ponavlja dok ne ostane samo jedan vojnik — **pobjednik**.

Na primjer, za `n = 7` vojnika i `k = 3`:
```
Vojnici u krugu: [1, 2, 3, 4, 5, 6, 7]

Krug 1: 1, 2, [3] → eliminiran 3
Krug 2: 4, 5, [6] → eliminiran 6
Krug 3: 7, 1, [2] → eliminiran 2
Krug 4: 4, 5, [7] → eliminiran 7
Krug 5: 1, 4, [5] → eliminiran 5
Krug 6: 1, [4] → eliminiran 4
Pobjednik: 1
```

Koristeći `CDLL` i metodu `rotate()` iz Zadatka 1., implementirajte funkciju `josephus_cdll(n, k)` koja simulira opisani postupak i ispisuje redoslijed eliminacije te na kraju vraća poziciju pobjednika.

````python
def josephus_cdll(n, k):

  soldiers = CDLL()
  cycle = 1

  for _ in range(1, n+1):
    soldiers.add_last(_)

  while soldiers.size > 1:

    soldiers.rotate(k-1)
    eliminated = soldiers.head.value

    current = soldiers.head
    next_in_line = []

    for _ in range(k-1):
      next_in_line.append(str(current.value))
      current = current.next

    survived = ", ".join(next_in_line)
    print(f"Cycle {cycle}: {survived}, [{eliminated}] -> eliminated {eliminated} ")

    soldiers.delete_first()
    cycle += 1

  print("----------------------")
  winner = soldiers.head.value
  print(f"The winner is: {winner}")



````
